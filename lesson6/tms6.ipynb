```python
#Task2
def bank(pin):
    if len(pin) == 4 and pin.isdigit():
        return True
    if len(pin) == 6 and pin.isdigit():
        return True
    return False

pin_1 = '132425'
pin_2  = '4g44'
pin_3  = '4444'

result_1 = bank(pin_1)
result_2 = bank(pin_2)
result_3 = bank(pin_3)

print(result_1, result_2, result_3)
```

```python
#Task4
def unique_in_order(lst:list):
    new_lst = []
    for q in range(len(lst)):
        if (q == len(lst) - 1 or lst[q] != lst[q + 1] ) and (lst[q] != lst[q - 1] or q == 0 ): 
            new_lst.append(lst[q])
    return new_lst

unique_in_order([6,1,1,1,2,2,3,1,5,6])
    
```

```python
#Task5
str_1 = 'hello world a ddddd ss BDBDB'

def f_min_word(str_1):
    words = str_1.split()
    min_word = words[0]
    for word in words:
        if len(word) < len(min_word):
            min_word = word
    return min_word

print(str_1)
print(f_min_word(str_1))
```

```python
#Task6
def f_digits(x : int):
    if x < 10:
        return x
    
    list_digits = []
    while x != 0:
        
        list_digits.append(x % 10)
        x = x // 10
    return list_digits

def sum_list(list_digits):
    sum = 0
    
    for i in range(len(list_digits)):
        sum += list_digits[i]
    return sum

def digital_root(n: int):
    while n >= 10:
        n = sum_list(f_digits(n))
    return n

n = 445
result = digital_root(n)
print(result)
```

```python
#Task8
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

print(is_prime(11))
print(is_prime(12))
```

```python
def check_sum_parity(numbers):
    total_sum = sum(numbers)
    if total_sum % 2 == 0:
        return ""  
    else:
        return "" 

print(check_sum_parity([2, 4, 6]))
print(check_sum_parity([1,2,3,7]))
```
