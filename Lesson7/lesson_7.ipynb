```python
class Money:
    def __init__(self, rubles: int, kopeeks: int):
        self.rubles = rubles
        self.kopeeks = kopeeks
        self._normalize()
    
    def _normalize(self):
        if self.kopeeks >= 100 or self.kopeeks < 0:
            self.rubles += self.kopeeks // 100
            self.kopeeks = self.kopeeks % 100
    
    @classmethod
    def from_float(cls, amount: float):
        rubles = int(amount)
        kopeeks = round((amount - rubles) * 100)
        return cls(rubles, kopeeks)
    
    def to_float(self):
        return self.rubles + self.kopeeks / 100
    
    def __str__(self):
        return f"{self.rubles},{self.kopeeks:02d}"
    
    def __repr__(self):
        return f"Money({self.rubles}, {self.kopeeks})"
    
    # Арифметические операции
    def __add__(self, other):
        if isinstance(other, Money):
            rubles = self.rubles + other.rubles
            kopeeks = self.kopeeks + other.kopeeks
            return Money(rubles, kopeeks)
        return NotImplemented
    
    def __sub__(self, other):
        if isinstance(other, Money):
            total_self = self.to_float()
            total_other = other.to_float()
            result = total_self - total_other
            return Money.from_float(result)
        return NotImplemented
    
    def __mul__(self, scalar):
        if isinstance(scalar, (int, float)):
            result = self.to_float() * scalar
            return Money.from_float(result)
        return NotImplemented
    
    def __truediv__(self, other):
        if isinstance(other, (int, float)):
            result = self.to_float() / other
            return Money.from_float(result)
        elif isinstance(other, Money):
            return self.to_float() / other.to_float()
        return NotImplemented
    
    def __rmul__(self, scalar):
        return self.__mul__(scalar)
    
    # Операции сравнения
    def __eq__(self, other):
        if isinstance(other, Money):
            return (self.rubles == other.rubles and 
                    self.kopeeks == other.kopeeks)
        return NotImplemented
    
    def __lt__(self, other):
        if isinstance(other, Money):
            return self.to_float() < other.to_float()
        return NotImplemented
    
    def __le__(self, other):
        if isinstance(other, Money):
            return self.to_float() <= other.to_float()
        return NotImplemented
    
    def __gt__(self, other):
        if isinstance(other, Money):
            return self.to_float() > other.to_float()
        return NotImplemented
    
    def __ge__(self, other):
        if isinstance(other, Money):
            return self.to_float() >= other.to_float()
        return NotImplemented


# Примеры использования
if __name__ == "__main__":
    # Создание объектов
    money1 = Money(100, 50)  # 100 рублей 50 копеек
    money2 = Money(50, 75)   # 50 рублей 75 копеек
    
    print(f"money1: {money1}")
    print(f"money2: {money2}")
    
    # Сложение
    sum_money = money1 + money2
    print(f"Сумма: {sum_money}")
    
    # Вычитание
    diff_money = money1 - money2
    print(f"Разность: {diff_money}")
    
    # Умножение на число
    multiplied = money1 * 1.5
    print(f"Умножение на 1.5: {multiplied}")
    
    # Деление на число
    divided = money1 / 2
    print(f"Деление на 2: {divided}")
    
    # Деление суммы на сумму
    ratio = money1 / money2
    print(f"money1 / money2 = {ratio:.2f}")
    
    # Сравнение
    print(f"money1 > money2: {money1 > money2}")
    print(f"money1 == money2: {money1 == money2}")
    
    # Создание из float
    money3 = Money.from_float(123.45)
    print(f"Из float 123.45: {money3}")
    
    # Проверка нормализации
    money4 = Money(100, 150)  # 100 рублей 150 копеек
    print(f"До нормализации: Money(100, 150)")
    print(f"После нормализации: {money4}")
```
