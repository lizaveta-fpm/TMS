3. Напишите функцию change(lst), которая принимает список и меняет местами его
 первый и последний элемент. В исходном
 списке минимум 2 элемента.
 lst = [1, 2, 3, 4, 5]
 def change(lst):
 change(lst)

```python
def change(lst):
    if len(lst) >= 2:
        lst[0], lst[-1] = lst[-1], lst[0]
    return lst

lst : list = [1, 2, 3, 4, 5]
print(lst)
change(lst)
print(lst)
```

 4. Николай задумался о поиске «бесполезного» числа на основании списка. Суть
 оного в следующем: он берет произвольный список чисел, находит самое
 большое из них, а затем делит его на длину списка. Студент пока не придумал,
 где может пригодиться подобное значение, но ищет у вас помощи в реализации
 такой функции useless(s).

```python
def useless(s):
    if not s:  
        return 0
    max_number = max(s)
    length = len(s)
    return max_number / length

print(useless([1, 5, 77]))        # 77 / 3 = 25.666...
print(useless([10, 20, 30, 40]))  # 40 / 4 = 10.0
print(useless([100]))             # 100 / 1 = 100.0
```

 5. На входе имеем список строк разной длины. Необходимо написать функцию
 all_eq(lst), которая вернет новый список из строк одинаковой длины. Длину
 итоговой строки определяем исходя из самой большой из них. Если конкретная
 строка короче самой длинной, дополнить ее нижними подчеркиваниями с
 правого края до требуемого количества символов.
 Расположение элементов начального списка не менять.
 ['крот', 'белка', 'выхухоль']->
 ['крот____', 'белка___', 'выхухоль' реши на питоне

```python
def all_eq(lst):
    if not lst:  # если список пустой
        return []
    
    max_length = max(len(item) for item in lst)
    
    # Дополняем все строки до максимальной длины
    result = []
    for item in lst:
        # Дополняем строку символами '_' справа
        padded_item = item.ljust(max_length, '_')
        result.append(padded_item)
    
    return result

animals = ['крот', 'белка', 'выхухоль']
result = all_eq(animals)
print(result)
```
