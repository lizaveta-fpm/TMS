```python
import kagglehub
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as seaborn

# Download latest version
path = kagglehub.dataset_download("prasad22/healthcare-dataset")

print("Path to dataset files:", path)
```

```python
import os
df_path = os.path.join(path, 'healthcare_dataset.csv')
df = pd.read_csv(df_path)
```

```python
df.columns
```

```python
df = df.drop(['Name', 'Date of Admission', 'Doctor', 'Hospital', 'Insurance Provider', 'Billing Amount', 'Room Number', 
      'Admission Type', 'Discharge Date'], axis = 1 )
```

```python
df['Gender'] = df['Gender'].map({'Female': 1, 'Male': 0})
df['Test Results'] = df['Test Results'].map({'Normal': 0, 'Inconclusive': 1, 'Abnormal': 2})
```

```python
Medical_Condition_unique = np.unique(df[['Medical Condition']])
Blood_Type_unique = np.unique(df[['Blood Type']])
Medication_unique = np.unique(df[['Medication']])
```

```python
for i in range(len(Medical_Condition_unique)):
    df[f'Medical_Condition_{Medical_Condition_unique[i]}'] = df['Medical Condition'].map({f'{Medical_Condition_unique[i]}': 1, '0': 0}).fillna(0)
for i in range(len(Blood_Type_unique)):
    df[f'Blood_Type_{Blood_Type_unique[i]}'] = df['Blood Type'].map({f'{Blood_Type_unique[i]}': 1, '0': 0}).fillna(0)
for i in range(len(Medication_unique)):
    df[f'Medication_{Medication_unique[i]}'] = df['Medication'].map({f'{Medication_unique[i]}': 1, '0': 0}).fillna(0)
```

```python
df = df.drop(['Medical Condition', 'Blood Type', 'Medication'], axis = 1 )
df
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso, LogisticRegression
from sklearn.model_selection import train_test_split
from sklearn import svm

list_col = df.drop(['Test Results'], axis=1).columns.to_list()

x = df.loc[:,list_col].to_numpy()
y = df.loc[:,['Test Results']].to_numpy().flatten()

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=42)
```

```python
log = LogisticRegression(max_iter=1000).fit(x_train, y_train)
reg = LinearRegression().fit(x_train, y_train)
vector_1 = svm.SVC(C = 0.1).fit(x_train, y_train)
vector_2 = svm.SVC(C = 10).fit(x_train, y_train)
```

```python
print(f'линейная регрессия: {round(reg.score(x_train, y_train), 2)} / {round(reg.score(x_test, y_test), 2)}')
print(f'логистическая регрессия: {round(log.score(x_train, y_train), 2)} / {round(log.score(x_test, y_test), 2)}')
print(f'метод опорных векторов с параметром С = 0.1: {round(vector_1.score(x_train, y_train), 2)} / {round(vector_1.score(x_test, y_test), 2)}')
print(f'метод опорных векторов с параметром С = 10: {round(vector_2.score(x_train, y_train), 2)} / {round(vector_2.score(x_test, y_test), 2)}')
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(random_state=42).fit(x_train, y_train)
print(f'Дерево решений: {round(dt.score(x_train, y_train), 3)} / {round(dt.score(x_test, y_test), 3)}')
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
param_grid = {
    'max_depth': [5, 10, 15, 20, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'criterion': ['gini', 'entropy']
}

grid_search = GridSearchCV(
    DecisionTreeClassifier(random_state=42),
    param_grid,
    cv=5,                   # 5-кратная кросс-валидация
    scoring='accuracy',
    n_jobs=-1,    # использовать все ядра CPU
    verbose=1     # показывать прогресс
)

grid_search.fit(x_train, y_train)

print(f'\nЛучшие параметры: {grid_search.best_params_}')
print(f'Лучшая кросс-валидационная точность: {round(grid_search.best_score_, 3)}')
```

```python
best_dt = grid_search.best_estimator_
train_score = best_dt.score(x_train, y_train)
test_score = best_dt.score(x_test, y_test)
print(f'Точность на обучении: {round(train_score, 3)}')
print(f'Точность на тесте: {round(test_score, 3)}')
```

```python
# Сравнение с дефолтными параметрами
default_tree = DecisionTreeClassifier(random_state=42)
default_tree.fit(x_train, y_train)
print(f"Default тест: {default_tree.score(x_test, y_test):.3f}")
```
