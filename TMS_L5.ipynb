```python
import random
i = random.randint(0, 100000)
print(i)
def rev_num(num):
    list_1 = []
    num_str = str(num)
    for a in range(1, len(num_str)+1):
        digit = (int(num) % 10 ** a ) // (10 ** (a-1))
        list_1.append(digit)
    return list_1           

    
result = rev_num(i)
result


```

```python
def reverse_digits(number):
    digits = []
    for digit in str(number):
        digits.append(int(digit))
    return digits[::-1]

print(reverse_digits(35231))  
```

```python
n = 4
matrix = []
for i in range(n):
    row = list(range(i*n + 1, (i+1)*n + 1) )
    if i % 2 == 1:
        row.reverse()
    matrix.append(row)
matrix    
```

```python
n = 4
matrix = []

for col in range(n):
    if col % 2 == 0: 
        #     
        column_numbers = list(range(col * n + 1, (col + 1) * n + 1))
    else: 
        #     
        column_numbers = list(range((col + 1) * n, col * n, -1))
    
    #      
    for i in range(n):
        if len(matrix) <= i:
            matrix.append([])
        matrix[i].append(column_numbers[i])

for row in matrix:
    print(row)
```
